# Question 2
* We created My Perceptron class for only 2 inputs. Extend this code for 3 inputs. Investigate the iris data set and choose 3 features to classify setosa and versicolor using your code. Notice that you cannot easily plot the decision boundary now since the data is 3-dimensional, but you can still compare the actual and the predicted labels to see how your algorithm is performing.

```python
# importing libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import a function for plotting decision boudaries
!pip install mlxtend
from mlxtend.plotting import plot_decision_regions
```

```python
# Reading Dataset
df = pd.read_csv("iris_dataset.csv")
```

```python
df
```

```python
df['species'].value_counts()
```

```python
# we are extraction setosa and versicolor data from the data set
df = df.iloc[:100]
df
```

```python
# 3 input features are sepal_lengtgh , Petal_length and petal_width. X is input variable
X = df[["sepal_length", "petal_length", "petal_width"]].to_numpy()

X
```

```python

class MyPerceptron(object):
    def __init__(self, eta = 0.5, epochs = 50): # epochs = 50 in order , learning rate= 0.5
        self.eta = eta
        self.epochs = epochs

    def fit(self, X, y):
        self.w1 = np.random.rand(1)
        self.w2 = np.random.rand(1)
        self.w3 = np.random.rand(1) # w1,w2 w3 are weights
        self.b = np.random.rand(1)

        self.errors = []

        for _ in range(self.epochs):
            errors = 0
            for xi, yi in zip(X, y):
                update = self.eta * (self.predict(xi) - yi)
                self.w1 = self.w1 - update*xi[0]
                self.w2 = self.w2 - update*xi[1]
                self.w3 = self.w3 - update*xi[2] # updating the weights
                self.b = self.b - update # updation of bias
                errors = errors + int(update != 0)
            if errors == 0:
                return self
            else:
                self.errors.append(errors)

        return self

    def weighted_sum(self, x):
        self.w = np.array([self.w1, self.w2, self.w3])
        return np.dot(x, self.w) + self.b

    def predict(self, x):
        return np.where(self.weighted_sum(x) > 0.0, 1, -1)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html
from sklearn.linear_model import Perceptron
```

```python
my_clf = MyPerceptron()
```

```python
# Call the fit method
my_clf.fit(X, y)
```

```python
y_pred = my_clf.predict(X)
y_pred
```

```python
y
```

```python
y_pred = y_pred.reshape(100)
y_pred
```

```python
print(y == y_pred)
# we are comparing the actual and predicted output values. we can clearly observe both values are equivalent.
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of test data
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', marker='o', label='Blue : setosa, Yellow : versicolor')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Width')

# Plot the decision boundary
xx, yy = np.meshgrid(np.arange(X[:, 0].min() - 1, X[:, 0].max() + 1, 0.02),
                     np.arange(X[:, 1].min() - 1, X[:, 1].max() + 1, 0.02))
zz = (-my_clf.w1 * xx - my_clf.w2 * yy - my_clf.b) / my_clf.w3 #This is the seperator boundary equation which divides w3 in order to plot my_clf which is MyPerceptron

ax.plot_surface(xx, yy, zz, alpha=0.3, cmap='coolwarm')
ax.view_init(1,20);

plt.title('Perceptron Model in 3D')
plt.legend()
plt.show()
```

```python

class MyPerceptron1(object):
    def __init__(self, eta = 0.5, epochs = 50):
        self.eta = eta
        self.epochs = epochs

    def fit(self, X, y):
        self.w = np.random.rand(X.shape[1])
        self.b = np.random.rand(1)

        self.errors = []

        for _ in range(self.epochs):
            errors = 0
            for xi, yi in zip(X, y):
                update = self.eta * (self.predict(xi) - yi)
                self.w=self.w - update*xi
                self.b = self.b - update
                errors = errors + int(update != 0)
            if errors == 0:
                return self
            else:
                self.errors.append(errors)

        return self

    def weighted_sum(self, x):
        return np.dot(x, self.w) + self.b

    def predict(self, x):
        return np.where(self.weighted_sum(x) > 0.0, 1, -1)
```
